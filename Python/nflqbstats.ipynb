```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from pandas import Series,DataFrame

# Set up empty data list
data = []

#Setup for multiple pages with same table structure

#year = 2015 

#for each year a file will be created

for year in xrange(2013,2016):
    
    for numb in ('1', '42', '82'):
        address = ('http://espn.go.com/nfl/statistics/player/_/stat/passing/sort/passingYards/year/' + str(year) + '/qualified/false/count/' + str(numb))

        #print address

        # Request content from web page
        result = requests.get(address)
        webcontent = result.content

        # Set as Beautiful Soup Object
        soup = BeautifulSoup(webcontent)

        # Go to the section of interest
        summary = soup.find("div",{'id':'my-players-table','class':'col-main'})

        # Find the tables in the HTML
        tables = summary.find_all('table')

        #print summary

        # Set rows as first indexed object in tables with a row
        rows = tables[0].findAll('tr')

        # now grab every HTML cell in every row
        for tr in rows:
            cols = tr.findAll('td')

            # Check to see if text is in the row
            for td in cols:
                text = td.find(text=True) 

                #print text,
                data.append(text)

                #removing unwanted header names from data
                chars = ['RK','PLAYER','TEAM','COMP','ATT','PCT','YDS','YDS/A','LONG','TD','INT','SACK','RATE','YDS/G']
                data = [list_data for list_data in data if list_data not in chars]

        rk = []
        player = [] 
        team = []
        comp = []
        att = []
        pct = []
        yds = []
        ydsa = []
        longyrd = []
        tds = []
        intercep = []
        sack = []
        rate = []
        ydsg = []
        yr = []

        index = 0

        for item in data:
            if index < len(data):

                rk.append(data[index + 0])
                player.append(data[index + 1])
                team.append(data[index + 2])
                comp.append(data[index + 3])
                att.append(data[index + 4])
                pct.append(data[index + 5])
                yds.append(data[index + 6])
                ydsa.append(data[index + 7])
                longyrd.append(data[index + 8])
                tds.append(data[index + 9])
                intercep.append(data[index + 10])
                sack.append(data[index + 11])
                rate.append(data[index + 12])
                ydsg.append(data[index + 13])
                yr.append(year) #adding a new column with the year data
            index = index + 14

        rk = Series(rk)
        player = Series(player)
        team = Series(team)
        comp = Series(comp)
        att = Series(att)
        pct = Series(pct)
        yds = Series(yds)
        ydsa = Series(ydsa)
        longyrd = Series(longyrd)
        tds = Series(tds)
        intercep = Series(intercep)
        sack = Series(sack)
        rate = Series(rate)
        ydsg = Series(ydsg)
        yr = Series(yr)

        nflstats_df = pd.concat([rk,player,team,comp,att,pct,yds,ydsa,longyrd,tds,intercep,sack,rate,ydsg,yr],axis = 1)

        #Give header
        nflstats_df.columns = ['RK','PLAYER','TEAM','COMP','ATT','PCT','YDS','YDS/A','LONG','TD','INT','SACK','RATE','YDS/G','YEAR']

        #Copying index value + 1 to RK column...RK column had blank values since multiple players had same data
        nflstats_df['RK'] = nflstats_df.index + 1

        #copying the data to csv and ignoring the index column
        nflstats_df.to_csv('nflstats' + str(year) + '.csv',index=False)
```
